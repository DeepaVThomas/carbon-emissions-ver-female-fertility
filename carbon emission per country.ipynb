```python
bins = np.linspace(min(df['2014']), max(df['2014']), 4)
group = ['low', 'medium', 'high']
df['last_bin'] = pd.cut(df['2014'], bins, labels=group, include_lowest=True)
df.tail(5)
```

```python
df.dropna(subset=['first_bin'], axis=0, inplace=True)
df.tail(10)
```

```python
n_df = df.loc[(df['first_bin'] == 'low') & (df['last_bin'] != 'low')]
n_df
```

```python
ax = n_df[['1961','2014']].plot(kind='bar', figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Country", fontsize=12)
ax.set_ylabel("Emission value", fontsize=12)
plt.show()
```

```python
bins = np.linspace(min(w_df['1961']), max(w_df['1961']), 4)
group = ['low', 'medium', 'high']
w_df['first'] = pd.cut(w_df['1961'], bins, labels=group, include_lowest=True)
w_df.head(5)
```

```python
n_df = w_df.loc[(w_df['first'] == 'high') & (w_df['latest'] == 'low')]
```

```python
ax = n_df[['1961','2014']].plot(kind='bar', figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Country", fontsize=12)
ax.set_ylabel("Child per woman", fontsize=12)
plt.show()
```

```python
df['1961'] = df['1961']/df['1961'].max()
df['2014'] = df['2014']/df['2014'].max()
y1 = np.array(df.loc[df['country'] == 'Luxembourg'])
print(y1)
```

```python
w_df['1961'] = w_df['1961']/w_df['1961'].max()
w_df['2014'] = w_df['2014']/w_df['2014'].max()
y2 = np.array(w_df.loc[w_df['country/year'] == 'Luxembourg'])
print(y2)
```

```python
yc = np.array([y1[0,1], y1[0,-1]])
print(yc)
```

```python
yw = np.array([y2[0,1], y2[0,-1]])
print(yw)
```

```python
x = np.array([1961, 2014])
print(x)
```

```python
plt.figure(figsize=(30,15))
plt.plot(x, yw, 'r.-', label='Child per woman')
plt.xticks(x)
plt.plot(x, yc, 'b.-', label='Carbon emission')
plt.xticks(x)
plt.legend()
plt.title('Children per woman vs Carbon emission in Luxembourg')
plt.grid()
plt.show()
```
